# ジェネレータの実験

```python
import numpy as np

def affine_parameter_generator():
    yield(1.0, 0.0)
    for t in 2**(0.5*np.arange(1, 6)):
        for phi in np.arange(0, 180, 72.0 / t):
            yield (t, phi)
```

```python
ag = affine_parameter_generator()
```

```python
for p in ag:
    print(p)
```
